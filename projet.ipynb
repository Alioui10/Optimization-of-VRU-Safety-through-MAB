```python
import pandas as pd
#import the data
data = pd.read_excel('ligne.xlsx')
#data = pd.read_excel('croisement.xlsx')
```

```python

# Separate vehicle data
vehicle_columns = ['time', 'id', 'x', 'y', 'angle', 'type', 'speed', 'pos', 'lane', 'slope']
vehicle_data = data[vehicle_columns]

# Separate pedestrian data
pedestrian_columns = {'time': 'time', 'id2': 'id_ped', 'x3': 'x_ped', 'y4': 'y_ped', 'angle5': 'angle_ped',
                      'speed6': 'speed_ped', 'pos7': 'pos_ped', 'edge': 'lane_ped', 'slope8': 'slope_ped'}
pedestrian_data = data[list(pedestrian_columns.keys())].rename(columns=pedestrian_columns)

# Display the first few rows of both datasets to ensure they are correctly separated and renamed
vehicle_data.head(), pedestrian_data.head()
```

```python
# Supprimer les lignes contenant des valeurs NaN
df_cars_cleaned = vehicle_data.dropna()
df_pedestrians_cleaned = pedestrian_data.dropna()
vehicle_data=df_cars_cleaned
pedestrian_data=df_pedestrians_cleaned
```

```python
import numpy as np
import math
unique_vehicles = vehicle_data['id'].unique()
rewards = {v: 0 for v in unique_vehicles}  # Tracking rewards for each vehicle
epsilon=0.1

def choose_vehicle(data, n, epsilon, flag):
    """ Randomly choose a vehicle to explore or exploit the best performing one. """
    if np.random.rand() < epsilon or flag:
        return np.random.choice(data, size=n, replace=False)
    return_list=[]
    i=0
    while len(return_list)<n:
      if best_vehicles[i] in list(data):
         return_list.append(best_vehicles[i])
      i+=1
    return return_list

def update_rewards(vehicle_id, reward):
    """ Update the rewards for a given vehicle and determine the best performing vehicle. """
    global best_vehicles
    rewards[vehicle_id] += reward
    # Update the best performing vehicle
    best_vehicles = sorted(rewards, key=rewards.get, reverse=True)

def calculate_distance_to_nearest_vru(vehicle, vr_data):
    """
    Calculate the minimum Euclidean distance from a given vehicle to the nearest VRU.

    Parameters:
        vehicle (pd.Series): A pandas Series containing the data of the current vehicle.
        vr_data (pd.DataFrame): A pandas DataFrame containing the data of all VRUs at the current time.

    Returns:
        float: The minimum distance to the nearest VRU.
    """
    # Extract the x and y coordinates of the current vehicle
    x_vehicle, y_vehicle = vehicle['x'], vehicle['y']

    # Calculate the Euclidean distances from the current vehicle to all VRUs
    distances = np.sqrt((vr_data['x_ped'] - x_vehicle)**2 + (vr_data['y_ped'] - y_vehicle)**2)

    # Return the minimum distance found
    return distances.min()

def calculate_distances(vehicle, vr_data):
    distances=[]
    for index, v in vehicle.iterrows():
      distances.append(calculate_distance_to_nearest_vru(v,vr_data))
    return distances

def evaluate_communication_success(distance,action):
    """
    Evaluate the communication success based on the distance and action taken.

    Parameters:
        distance (float): The distance to the nearest VRU.
        action (int): The action taken, 0 or 1, where 1 indicates communication was initiated.
        threshold (float): Distance threshold for considering communication necessary and effective.

    Returns:
        int: The numerical reward for the action taken based on the specified criteria.
    """
    return max(0,20 - distance)

def taux_couverture(vehicle_d, pedestrian):
    # Initialize a list to store the results
    taux = [0]*len(pedestrian)
    i=0
    # Iterate over each vehicle
    for index_ped, pedestrian in pedestrian.iterrows():

        # Compare distance with each pedestrian
        for index, vehicle in vehicle_d.iterrows():
            distance = np.sqrt((vehicle['x'] - pedestrian['x_ped'])**2 + (vehicle['y'] - pedestrian['y_ped'])**2)
            if distance<20 :
               taux[i]=1
               break
        i+=1
    if len(taux)==0:
      return 1

    return sum(taux)/len(taux)

def are_pedestrians_seen(vehicles, pedestrians, visibility_distance):
    """
    Dtermine si les pitons sont vus par les vhicules en fonction de la distance de visibilit.

    :param vehicles: Liste de tuples (x, y) reprsentant les positions des vhicules.
    :param pedestrians: Liste de tuples (x, y) reprsentant les positions des pitons.
    :param visibility_distance: Distance maximale  laquelle un vhicule peut voir un piton.
    :return: Liste de boolens indiquant si chaque piton est vu par au moins un vhicule.
    """
    seen = [False] * len(pedestrians)
    nombre_vehicule_vu=[]*len(pedestrians)
    for i, ped in pedestrians.iterrows():
        sum=0
        for index, veh in vehicles.iterrows():

            distance = np.sqrt((veh['x'] - ped['x_ped']) ** 2 + (veh['y'] - ped['y_ped']) **2)
            if distance <= visibility_distance:
                sum+=1
        nombre_vehicule_vu.append(sum)
    return nombre_vehicule_vu





cars=[]
taux={}   #taux optimal
for time in vehicle_data['time'].unique():
      current_data = vehicle_data[vehicle_data['time'] == time]
      current_pedestrian_data=pedestrian_data[pedestrian_data['time'] == time]
      taux[time]=taux_couverture(current_data,current_pedestrian_data)
      cars_seen=are_pedestrians_seen(current_data, current_pedestrian_data, 20)
      cars+=cars_seen
j=70
cars_algorithm=[]
taux_3={}
flag=True

#effectuer l'algorithme et calculer le taux et les critres aprs l'algorithme
for time in vehicle_data['time'].unique():
    current_data = vehicle_data[vehicle_data['time'] == time]
    current_pedestrian_data=pedestrian_data[pedestrian_data['time'] == time]
    vehicle_ids=[]
    n=math.floor(len(current_data)*j/100)
    vehicle_ids = choose_vehicle(current_data['id'],n,epsilon,flag)
    flag=False
    vehicle = current_data[current_data['id'].isin(vehicle_ids)]
    distances = calculate_distances(vehicle, current_pedestrian_data)
    for i in range(0,n):
      reward = evaluate_communication_success(distances[i], 1)-(vehicle['speed'].iloc[i])
      update_rewards(vehicle_ids[i], reward)
    taux_3[time]=taux_couverture(vehicle,current_pedestrian_data)
    cars_seen=are_pedestrians_seen(vehicle, current_pedestrian_data, 20)
    cars_algorithm+=cars_seen
```
